```python
import serial
import time
import math
```

```python
# 0.1mm => 1deg

MAX_SPEED = 250.0

FOCAL_LENGTH = 500.0
SENSOR_WIDTH = 36.0
SENSOR_HEIGHT = 24.0
OVERLAP = 0.25

# In degrees
PANORAMA_WIDTH = 180.0
PANORAMA_HEIGHT = 30.0
```

```python
aov_h = 2*math.atan(SENSOR_WIDTH/(2*FOCAL_LENGTH))*180/math.pi
aov_v = 2*math.atan(SENSOR_HEIGHT/(2*FOCAL_LENGTH))*180/math.pi

overlap_h = aov_h * OVERLAP
overlap_v = aov_v * OVERLAP

image_step_h = aov_h - overlap_h
image_step_v = aov_v - overlap_v

image_step_h
```

```python
def send_gcode(gcode):
    ser.write(str.encode(gcode) + str.encode('\n'))

def init_grbl():
    send_gcode("G0 G54 G17 G21 G90 G94 M5 M9 T0 F0")

def set_zero():
    send_gcode("G10 P0 L20 X0 Y0")

def move_to_zero(axis='X'):
    send_gcode(f"G1 G90 {axis}0 F250")

def move_deg(degrees,speed=200,axis='X'): # axis= 'X' or 'Y'
    if speed > 250:
        speed = 250
    send_gcode(f"G1 G91 {axis}{degrees/100.0} F{speed}")

ser = serial.Serial(port="COM4",baudrate=115200)
ser.write(str.encode("\r\n\r\n"))
time.sleep(2)
ser.flushInput()


init_grbl()
set_zero()

for i in range (int(PANORAMA_HEIGHT/image_step_v)):
    for i in range (int(PANORAMA_WIDTH/image_step_h)):
        move_deg(image_step_h,200,axis='X')
        time.sleep(1)
        print(f"taking picture: {i}")
    move_to_zero(axis='X')
    move_deg(image_step_v,200,axis='Y')
move_to_zero(axis='X')
move_to_zero(axis='Y')
    

ser.close()

```
